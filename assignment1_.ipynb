```python
def max_meetings(meetings):
    meetings.sort(key=lambda meeting: meeting[1])
    meeting_count = 0
    last_meeting_end = -1
    for start_time, end_time in meetings:
        if start_time >= last_meeting_end:
            meeting_count += 1
            last_meeting_end = end_time
    return meeting_count
T = int(input())
for i in range(T):
    N = int(input())
    meetings = []
    for j in range(N):
        start, end = map(int, input().split())
        meetings.append((start, end))
    print(max_meetings(meetings))
```

```python
def first_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            high = mid - 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans
def last_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            low = mid + 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans
T = int(input())
for _ in range(T):
    N, Q = map(int, input().split())
    arr = list(map(int, input().split()))

    query_values = list(map(int, input().split()))
    for x in query_values:
        first = first_occurrence(arr, x)
        if first == -1:
            print(0)
        else:
            last = last_occurrence(arr, x)
            print(last - first + 1)
```
