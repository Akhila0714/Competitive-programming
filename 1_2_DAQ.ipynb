```python
def max_crossing_sum(arr, left, mid, right):
    left_sum = float('-inf')
    sum_val = 0
    for i in range(mid, left - 1, -1):
        sum_val += arr[i]
        left_sum = max(left_sum, sum_val)
        right_sum = float('-inf')
    sum_val = 0
    for i in range(mid + 1, right + 1):
        sum_val += arr[i]
        right_sum = max(right_sum, sum_val)
    return left_sum + right_sum
def max_subarray_sum(arr, left, right):
    if left == right:
        return arr[left]
    mid = (left + right) // 2
    left_max = max_subarray_sum(arr, left, mid)
    right_max = max_subarray_sum(arr, mid + 1, right)
    cross_max = max_crossing_sum(arr, left, mid, right)
    return max(left_max, right_max, cross_max)
t = int(input())
for i in range(t):
    n = int(input())
    arr = list(map(int, input().split()))
    result = max_subarray_sum(arr, 0, n - 1)
    print(result)
```
